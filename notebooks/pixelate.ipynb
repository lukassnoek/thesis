# Pixelate image for cover

```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from PIL import Image
```

```python
img = Image.open("to_pixelate.png")
#img = gaussian_filter(img, sigma=50)
#img = Image.fromarray(img)
small_pix = (30, 42)
img_s = img.resize(small_pix,resample=Image.BILINEAR)
img_s = np.array(img_s)

PIX = 40
for _ in range(PIX):
    x = np.random.choice(range(small_pix[1]))
    y = np.random.choice(range(small_pix[0]))
    img_s[x, y, :] = np.random.choice(range(5, 35))

img_s = Image.fromarray(img_s)
img = img_s.resize(img.size, resample=Image.NEAREST)

img_np = np.array(img)
img_np = img_np / img_np.max()

plt.figure(figsize=(20.08, 28.35))
plt.imshow(img_np)
plt.axis('off')
plt.savefig('yhat_pixelate.png', dpi=100)
```
